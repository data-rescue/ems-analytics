```python
import matplotlib.pyplot as plt
import numpy as np

x_min = 0.0
x_max = 12.0

x = np.linspace(x_min, x_max, 100)
y = np.linspace(x_min, x_max, 100)

plt.plot(x,y, color='black')

plt.xlim(0,12)
plt.ylim(0,12)

plt.title("How to extract code color from a seaborn palette ?",fontsize=10)

ptx = np.linspace(0, 2, 10)
pty = ptx
plt.fill_between(ptx, pty, color='#dbe9f6', alpha=None)

ptx = np.linspace(2, 4, 10)
pty = ptx
plt.fill_between(ptx, pty, color='#bad6eb', alpha=None)

ptx = np.linspace(4, 6, 10)
pty = ptx
plt.fill_between(ptx, pty, color='#89bedc', alpha=None)

ptx = np.linspace(6, 8, 10)
pty = ptx
plt.fill_between(ptx, pty, color='#539ecd', alpha=None)

ptx = np.linspace(8, 10, 10)
pty = ptx
plt.fill_between(ptx, pty, color='#2b7bba', alpha=None)

ptx = np.linspace(10, 12, 10)
pty = ptx
plt.fill_between(ptx, pty, color='#0b559f', alpha=None)

##plt.savefig("extract_seaborn_palette.png")
plt.show()
```
